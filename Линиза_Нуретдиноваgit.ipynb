1. Подсчет частоты слов: Напишите функцию, которая принимает строку и возвращает словарь, где ключи — это слова, а значения — их частота в строке.

```python
from collections import defaultdict
import re

def dictCount(text):
    words = re.findall(r'\w+', text.lower())
    frequency = defaultdict(int)
    for words in words:
        frequency[words] += 1
    return dict(frequency)

text = input()
print(dictCount(text))
```

2. Инвертирование словаря: Напишите функцию, которая принимает словарь и возвращает новый словарь, где ключи и значения поменяны местами. Если значения не уникальны, объедините ключи в список.

```python
def inverted_dict(original):
  inverted = {}
  for key, value in original.items():
    if value not in inverted:
      inverted[value] = key
    else:
      if not isinstance(inverted[value], list):
        inverted[value] = [inverted[value]]
      inverted[value].append(key)
  return inverted

dict1 = {1: "me", 2: "you", 3: "us"}
print(inverted_dict(dict1))


```

# 3. Слияние словарей: Напишите функцию, которая принимает два словаря и возвращает новый словарь, содержащий все ключи из обоих словарей. Если ключи совпадают, значения должны быть суммированы.

```python
def merge_dicts(dict1, dict2):
    merged = dict1.copy()
    for key, value in dict2.items():
        if key in merged:
            merged[key] += value
        else:
            merged[key] = value
    return merged

# Пример использования
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}

result = merge_dicts(dict1, dict2)
print(result)


```

4. Фильтрация по значению: Напишите функцию, которая принимает словарь и пороговое значение, и возвращает новый словарь, содержащий только те пары ключ-значение, где значение больше порогового.

```python
def filter_dict(input_dict, threshold):
    return {key: value for key, value in input_dict.items() if value > threshold}

my_dict = {'a': 10, 'b': 5, 'c': 15, 'd': 3, 'e': 20}
value1 = 7

filtered_dict = filter_dict(my_dict, value1)
print(filtered_dict)
```

5. Максимальное значение: Напишите функцию, которая находит ключ с максимальным значением в словаре.

```python
def find_max_key(dictionary):
    if not dictionary:
        return None
    return max(dictionary, key=dictionary.get)
my_dict = {'a': 10, 'b': 25, 'c': 15, 'd': 30, 'e': 5}
max_key = find_max_key(my_dict)
print(max_key)
```

6. Удаление ключей: Напишите функцию, которая принимает словарь и список ключей, и удаляет из словаря все указанные ключи.

```python
def remove_keys(dictionary, keys_to_remove):
    for key in keys_to_remove:
        dictionary.pop(key, None)
    return dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
keys = ['b', 'd', 'x']
result = remove_keys(my_dict, keys)
print(result)
```

7. Группировка по длине ключей: Напишите функцию, которая группирует ключи словаря по их длине и возвращает словарь, где ключи — это длины, а значения — списки ключей соответствующей длины.

```python
def group_keys_by_length(dictionary):
    result = {}
    for key in dictionary.keys():
        length = len(str(key))
        if length not in result:
            result[length] = []
        result[length].append(key)
    return result

my_dict = {
    'Liniza': 1,
    'Tasya': 2,
    'Zlata': 3,
    42: 4,
    'orange': 5,
    'pear': 6,
    (1, 2): 7
}

grouped = group_keys_by_length(my_dict)
print(grouped)
```

10. Глубокая копия словаря: Напишите функцию, которая создает глубокую копию словаря, включая все вложенные структуры данных.

```python
import copy

def deep_copy_dict(original_dict):
    return copy.deepcopy(original_dict)

original = {
    'birds': 1,
    'crow': [2, 3, 4],
    'owl': {'bird': 10, 'animal': 20},
    'pigeon': {'nested': {'deep': [5, 6]}}
}

copied = deep_copy_dict(original)
copied['crow'].append(5)
copied['owl']['bird'] = 100
copied['pigeon']['nested']['deep'].append(7)

print("Оригинал:", original)
print("Копия:", copied)
```

4. Объединение нескольких множеств: Напишите функцию, которая принимает список множеств и возвращает их объединение.

```python
def union_of_sets(list_of_sets):
    result = set()
    for s in list_of_sets:
        result.update(s)
    return result
sets = [{1, 2, 3}, {3, 4, 5}, {5, 6, 7}]
print(union_of_sets(sets))
```

5. Пересечение нескольких множеств: Напишите функцию, которая принимает список множеств и возвращает их пересечение.

```python
def intersection(list_of_sets):
    if not list_of_sets:
        return set()
    result = list_of_sets[0].copy()
    for s in list_of_sets[1:]:
        result.intersection_update(s)
    return result

sets = [{1, 2, 3}, {2, 3, 4}, {3, 4, 5}]
print(intersection(sets))
```

6. Разбиение множества: Напишите функцию, которая разбивает множество на два: одно с четными числами, другое с нечетными.

```python
def split_set(input_set):
    even = {x for x in input_set if x % 2 == 0}
    odd = {x for x in input_set if x % 2 != 0}
    return even, odd

numbers = {1, 2, 3, 4, 5, 6}
even, odd = split_set(numbers)
print(even)
print(odd)
```

7. Фильтрация по условию: Напишите функцию, которая фильтрует множество, оставляя только те элементы, которые удовлетворяют заданному условию (например, больше 10).

```python
def filter_set(input_set, condition):
    return {x for x in input_set if condition(x)}

numbers = {5, 10, 15, 20, 25}
filtered = filter_set(numbers, lambda x: x > 10)
print(filtered)
```

8. Генерация множества: Напишите функцию, которая генерирует множество из n случайных чисел в заданном диапазоне.

```python
import random

def generate_random_set(n, start, end):
    return {random.randint(start, end) for _ in range(n)}

random_set = generate_random_set(5, 1, 10)
print(random_set)
```

10. Удаление случайного элемента: Напишите функцию, которая удаляет случайный элемент из множества.

```python
import random

def remove_random_element(input_set):
    if not input_set:
        return set()
    element = random.choice(list(input_set))
    input_set.remove(element)
    return input_set

numbers = {10, 20, 30, 40, 50}
print(remove_random_element(numbers))
```

2.Развертывание вложенного кортежа: Напишите функцию, которая разворачивает вложенный кортеж в плоский кортеж.

```python
def flatten_tuple(nested):
    return tuple(item for sublist in nested for item in (sublist if isinstance(sublist, tuple) else (sublist,)))

nested = ((1, (2, 3)), (4, (5, 6)))
print(flatten_tuple(nested))
```

7. Кортеж с палиндромами: Напишите функцию, которая создает кортеж, содержащий только строки-палиндромы из заданного списка строк. напиши код отдельно для каждого задания и чтобы в коде не было коментариев

```python
def is_palindrome(s):
    return s == s[::-1]

def palindrome_tuple(strings):
    return tuple(s for s in strings if is_palindrome(s))

strings = ('madam', 'hello', 'level', 'world')
print(palindrome_tuple(strings))
```

8. Группировка элементов: Напишите функцию, которая группирует элементы кортежа по их типам (например, int, str, float) и возвращает словарь, где ключи — это типы, а значения — списки элементов.

```python
def group_by_type(input_tuple):
    result = {}
    for item in input_tuple:
        item_type = type(item).__name__
        if item_type not in result:
            result[item_type] = []
        result[item_type].append(item)
    return result

my_tuple = (1, 2.5, 'hello', 3, 'world', 4.2, True)
print(group_by_type(my_tuple))
```
